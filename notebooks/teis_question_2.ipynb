```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
with open('../data/teis_final_data.pickle', 'rb') as file:
    teis = pickle.load(file)
```

```python
teis.columns = [
    col.strip().lower().replace(' ', '_').replace('.', '_') 
    for col in 
    teis.columns]
teis.dtypes
```

```python
total_referrals = teis.groupby(['referral_source_type_name'])['successful'].count()
total_referrals
```

```python
total_referrals_successful = teis.groupby(['referral_source_type_name'])['successful'].sum()
total_referrals_successful
```

```python
tot_ref = pd.merge(total_referrals, total_referrals_successful, on='referral_source_type_name', how='left').rename(columns = {'successful_x' : 'total_referrals', 'successful_y' : 'total_successful_referrals'})
tot_ref
```

```python
tot_ref['percentage_success'] = (tot_ref['total_successful_referrals']/tot_ref['total_referrals']) * 100
tot_ref
```
